```python
import re
import time
import requests as r
from bs4 import BeautifulSoup as bs
import concurrent.futures
```

```python
def worker_item(items):
    domain = 'https://www.popdaily.com.tw/article/category/'
    for i in items:
        counter = 2
        switch = True
        p_url = 'https://www.popdaily.com.tw/article/category/{}'.format(i)
        p_resp = r.get(p_url)
        p_soup = bs(p_resp.text,'html5lib')
        urls.extend([x.get('href') for x in p_soup.select('div.container.category-page > div.w60 > a') \
                     if x.get('href').startswith('/article/')])
        print([x.get('href') for x in p_soup.select('div.container.category-page > div.w60 > a') \
                     if x.get('href').startswith('/article/')])
        pages = p_soup.select(".box-btns")
        while switch:
            if pages[-1].text == '下一頁':
                print(counter)
                p_url_2 = p_url+'/page/{}'.format(counter)
                p_resp_2 = r.get(p_url_2)
                p_soup_2 = bs(p_resp_2.text,'html5lib')
                urls.extend([x.get('href') for x in p_soup_2.select('div.container.category-page > div.w60 > a') \
                     if x.get('href').startswith('/article/')])
                counter += 1
                pages = p_soup_2.select(".box-btns")
                print([x.get('href') for x in p_soup_2.select('div.container.category-page > div.w60 > a') \
                     if x.get('href').startswith('/article/')])
            else:
                switch = False
```

```python
def worker(urls):
    counter = 0
    a_url = 'https://www.popdaily.com.tw'+urls  
    a_resp = r.get(a_url)
    a_soup = bs(a_resp.text,'html5lib')
    try:
        
        title = a_soup.select_one("article > div.article-header > h1.entry-title").text
        article = [i.text for i in a_soup.select("article > div.article-content > p")]
        with open('/home/user/popdaily2{}.txt'.format(urls),'w') as w:
            [w.write(i) for i in title]
            w.write('\n')
            [w.write(i) for i in article]
        counter += 1    
        print(counter)
    except:
        except_list.append(a_url)
```

```python
tStart = time.time()
urls = []
items = ['beauty','style','Tour','yummy','Japan','kimchi','life','love']
worker_item(items)
with open('/home/user/popdaily2/urls.txt','w') as f:
    for i in urls:
        f.write(i)
        f.write('\n')
tEnd = time.time()
print(tEnd - tStart)
```

```python
tStart = time.time()
except_list = []
with concurrent.futures.ThreadPoolExecutor(80) as executor:
    executor.map(worker, urls)
with open('/home/user/popdaily2/except_list.txt','w') as f:
    for i in except_list:
        f.write(i)
        f.write('\n')
tEnd = time.time()
print(tEnd - tStart)
```
